# 109 — Minimum Viable Benchmark

## Smallest benchmark for pruning-RML

This notebook defines the smallest benchmark capable of testing:

- residue hypotheses
- trigger hypotheses
- revision hypotheses

Outputs:

```text
results/109_minimum_viable_benchmark.csv
figures/109_minimum_viable_benchmark.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()
for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["baseline accuracy", 5, "required"],
    ["post-pruning accuracy", 5, "required"],
    ["residue retention", 5, "required"],
    ["trigger correctness", 4, "recommended"],
    ["revision cost", 5, "required"],
    ["compute budget", 5, "required"],
]

df = pd.DataFrame(
    rows,
    columns=["metric","priority","status"]
)

df
````

````python
csv_path = repo / "results" / "109_minimum_viable_benchmark.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "109_minimum_viable_benchmark.png"

ax = df.plot.bar(
    x="metric",
    y="priority",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Minimum viable benchmark")
ax.set_ylabel("Priority")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````

## Result

This notebook defines the smallest benchmark needed to move beyond toy examples.
