# 53 — Revision Lifecycle

## Multiple cycles of revision

Notebook 47 examined residue retention.

Notebook 53 asks:

> How do residues, triggers, budgets, and policies interact across repeated revision cycles?

Outputs:

```text
results/53_revision_lifecycle.csv
figures/53_revision_lifecycle.png
```

````python
from pathlib import Path
import os, pandas as pd, matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    [1, 5.0, 1.0, 2.0, "preserve_and_refine"],
    [2, 4.3, 1.4, 2.0, "preserve_and_refine"],
    [3, 3.7, 2.0, 1.0, "targeted_revision"],
    [4, 2.9, 3.0, 1.0, "targeted_revision"],
    [5, 1.8, 4.0, 0.0, "major_restructure"],
]

df = pd.DataFrame(
    rows,
    columns=[
        "cycle",
        "residue_strength",
        "trigger_pressure",
        "budget_headroom",
        "policy",
    ],
)

df["lifecycle_score"] = (
    df["residue_strength"]
    + df["budget_headroom"]
    - df["trigger_pressure"]
)

df
````

````python
csv_path = repo / "results" / "53_revision_lifecycle.csv"
df.to_csv(csv_path, index=False)

print(csv_path)
````

````python
fig_path = repo / "figures" / "53_revision_lifecycle.png"

ax = df.plot(
    x="cycle",
    y=["residue_strength","trigger_pressure","lifecycle_score"],
    figsize=(9,5),
)

ax.set_title("Revision lifecycle")
ax.set_ylabel("Score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````

## Result

Notebook 53 treats revision as a lifecycle rather than a single event.
