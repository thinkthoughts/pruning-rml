## Which residues persist over time?

Notebook 43 compared policies.

Notebook 47 asks:

> Which capabilities are most likely to survive repeated pruning and revision?

Outputs:

```text
results/47_residue_retention.csv
figures/47_residue_retention.png
```

````python
from pathlib import Path
import os, sys, subprocess

REPO_URL = "https://github.com/thinkthoughts/pruning-rml.git"
REPO_NAME = "pruning-rml"

cwd = Path.cwd().resolve()

if cwd.name == REPO_NAME:
    repo = cwd
elif Path("/content/pruning-rml").exists():
    repo = Path("/content/pruning-rml")
elif (cwd / REPO_NAME).exists():
    repo = cwd / REPO_NAME
else:
    target = Path("/content") if Path("/content").exists() else cwd
    os.chdir(target)
    subprocess.run(["git","clone",REPO_URL], check=True)
    repo = target / REPO_NAME

os.chdir(repo)
````

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

def ensure_dirs(root, names=("results","figures","reports")):
    root = Path(root)
    out={}
    for n in names:
        p=root/n
        p.mkdir(parents=True, exist_ok=True)
        out[n]=p
    return out

outputs = ensure_dirs(repo)
results = outputs["results"]
figures = outputs["figures"]
````

````python
rows = [
    ["attention structure", 0.92],
    ["token patterns", 0.88],
    ["feature circuits", 0.79],
    ["task heuristics", 0.73],
    ["memorized details", 0.41],
]

df = pd.DataFrame(rows, columns=["residue","retention_score"])
df
````

````python
csv_path = results / "47_residue_retention.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = figures / "47_residue_retention.png"

ax = df.plot.bar(
    x="residue",
    y="retention_score",
    legend=False,
    figsize=(9,5),
)

ax.set_title("Residue retention")
ax.set_ylabel("Retention score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
