## Architecture for continuous revision

Notebook 89 examined governance.

Notebook 97 asks:

> What architecture supports continuous revision without frequent reset?

Outputs:

```text
results/97_revision_architecture.csv
figures/97_revision_architecture.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["full reset pipeline", 0.40, 0.90, 0.50],
    ["periodic pruning", 0.75, 0.70, 0.75],
    ["trigger-driven revision", 0.88, 0.82, 0.85],
    ["continuous revision architecture", 0.95, 0.88, 0.92],
]

df = pd.DataFrame(
    rows,
    columns=[
        "architecture",
        "adaptability",
        "efficiency",
        "stability",
    ],
)

df["architecture_score"] = (
    df["adaptability"]
    + df["efficiency"]
    + df["stability"]
) / 3

df
````

````python
csv_path = repo / "results" / "97_revision_architecture.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "97_revision_architecture.png"

ax = df.plot.bar(
    x="architecture",
    y="architecture_score",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Revision architecture")
ax.set_ylabel("Architecture score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
