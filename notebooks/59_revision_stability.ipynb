# 59 — Revision Stability

## Which revision strategies remain stable?

Notebook 53 examined revision lifecycles.

Notebook 59 asks:

> Which strategies maintain capability while resisting drift across many cycles?

Outputs:

```text
results/59_revision_stability.csv
figures/59_revision_stability.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results", "figures", "reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["preserve_and_refine", 0.91, 0.88, 0.90],
    ["targeted_revision", 0.84, 0.82, 0.83],
    ["structured_pruning", 0.80, 0.79, 0.81],
    ["aggressive_pruning", 0.68, 0.72, 0.65],
    ["major_restructure", 0.61, 0.58, 0.60],
]

df = pd.DataFrame(
    rows,
    columns=[
        "strategy",
        "capability_retention",
        "drift_resistance",
        "stability",
    ],
)

df["overall_stability"] = (
    df["capability_retention"]
    + df["drift_resistance"]
    + df["stability"]
) / 3

df
````

````python
csv_path = repo / "results" / "59_revision_stability.csv"
df.to_csv(csv_path, index=False)

print(csv_path)
````

````python
fig_path = repo / "figures" / "59_revision_stability.png"

ax = df.plot.bar(
    x="strategy",
    y="overall_stability",
    legend=False,
    figsize=(9,5),
)

ax.set_title("Revision stability")
ax.set_ylabel("Overall stability")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
