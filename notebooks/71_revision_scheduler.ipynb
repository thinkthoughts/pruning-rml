# 71 — Revision Scheduler

## When should revision events occur?

Notebook 67 defined a revision boundary.

Notebook 71 asks:

> How should revision actions be scheduled as residue value and noise pressure evolve?

Outputs:

```text
results/71_revision_scheduler.csv
figures/71_revision_scheduler.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results", "figures", "reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    [1, 4.5, 0.5],
    [2, 4.0, 1.0],
    [3, 3.4, 1.8],
    [4, 2.8, 2.5],
    [5, 2.2, 3.1],
    [6, 1.7, 3.8],
]

df = pd.DataFrame(
    rows,
    columns=[
        "cycle",
        "residue_value",
        "noise_pressure",
    ],
)

df["boundary_score"] = df["residue_value"] - df["noise_pressure"]

def schedule_action(score):
    if score > 2:
        return "observe"
    if score > 0:
        return "prepare_revision"
    return "execute_revision"

df["scheduled_action"] = df["boundary_score"].apply(schedule_action)

df
````

````python
csv_path = repo / "results" / "71_revision_scheduler.csv"
df.to_csv(csv_path, index=False)

print(csv_path)
````

````python
fig_path = repo / "figures" / "71_revision_scheduler.png"

ax = df.plot(
    x="cycle",
    y=["residue_value","noise_pressure","boundary_score"],
    figsize=(9,5),
)

ax.set_title("Revision scheduler")
ax.set_ylabel("Score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````

## Result

Notebook 71 converts boundary detection into scheduling.
