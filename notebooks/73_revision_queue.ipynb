# 73 — Revision Queue

## Prioritizing multiple revision candidates

Notebook 71 scheduled revisions.

Notebook 73 asks:

> If several revision candidates exist simultaneously, which should be handled first?

Outputs:

```text
results/73_revision_queue.csv
figures/73_revision_queue.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["attention drift", 5, 4, 2],
    ["token inefficiency", 3, 2, 4],
    ["feature redundancy", 4, 3, 3],
    ["stale residue cluster", 5, 5, 1],
    ["budget hotspot", 2, 1, 5],
]

df = pd.DataFrame(
    rows,
    columns=[
        "candidate",
        "impact",
        "urgency",
        "budget_benefit",
    ],
)

df["queue_score"] = (
    df["impact"]
    + df["urgency"]
    + df["budget_benefit"]
)

df = df.sort_values("queue_score", ascending=False).reset_index(drop=True)
df["queue_position"] = df.index + 1

df
````

````python
csv_path = repo / "results" / "73_revision_queue.csv"
df.to_csv(csv_path, index=False)

print(csv_path)
````

````python
fig_path = repo / "figures" / "73_revision_queue.png"

ax = df.plot.bar(
    x="candidate",
    y="queue_score",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Revision queue priority")
ax.set_ylabel("Queue score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````

## Result

Notebook 73 converts revision scheduling into queue ordering.
