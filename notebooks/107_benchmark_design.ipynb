# 107 — Benchmark Design

## Designing a pruning-RML benchmark

This notebook proposes a benchmark structure for evaluating pruning-RML ideas.

Outputs:

```text
results/107_benchmark_design.csv
figures/107_benchmark_design.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()
for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["residue retention", 5, "core"],
    ["trigger accuracy", 5, "core"],
    ["queue effectiveness", 4, "secondary"],
    ["policy selection", 4, "secondary"],
    ["revision stability", 5, "core"],
    ["compute efficiency", 5, "core"],
]

df = pd.DataFrame(
    rows,
    columns=["benchmark_component","priority","tier"]
)

df
````

````python
csv_path = repo / "results" / "107_benchmark_design.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "107_benchmark_design.png"

ax = df.plot.bar(
    x="benchmark_component",
    y="priority",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Benchmark design priorities")
ax.set_ylabel("Priority")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
