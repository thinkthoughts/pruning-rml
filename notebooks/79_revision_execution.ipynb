## Measuring revision success

Notebook 73 prioritized revisions.

Notebook 79 asks:

> Once a revision is executed, how should success be measured?

Outputs:

```text
results/79_revision_execution.csv
figures/79_revision_execution.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["attention drift revision", 0.92, 0.81, 0.88],
    ["feature redundancy revision", 0.87, 0.84, 0.82],
    ["token inefficiency revision", 0.83, 0.91, 0.79],
    ["stale residue revision", 0.78, 0.76, 0.85],
]

df = pd.DataFrame(
    rows,
    columns=[
        "revision",
        "capability_retention",
        "efficiency_gain",
        "stability_after_revision",
    ],
)

df["execution_score"] = (
    df["capability_retention"]
    + df["efficiency_gain"]
    + df["stability_after_revision"]
) / 3

df
````

````python
csv_path = repo / "results" / "79_revision_execution.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "79_revision_execution.png"

ax = df.plot.bar(
    x="revision",
    y="execution_score",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Revision execution score")
ax.set_ylabel("Execution score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
