# 83 — Revision Feedback

## Learning from revision outcomes

Notebook 79 measured execution success.

Notebook 83 asks:

> How should revision outcomes influence future triggers, queues, and policies?

Outputs:

```text
results/83_revision_feedback.csv
figures/83_revision_feedback.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["high success revision", 0.90, 0.20, 0.15],
    ["moderate success revision", 0.75, 0.40, 0.35],
    ["mixed outcome revision", 0.60, 0.55, 0.50],
    ["low success revision", 0.40, 0.80, 0.75],
]

df = pd.DataFrame(
    rows,
    columns=[
        "revision_outcome",
        "execution_score",
        "future_trigger_pressure",
        "future_queue_priority",
    ],
)

df["feedback_score"] = (
    df["execution_score"]
    - df["future_trigger_pressure"]
)

df
````

````python
csv_path = repo / "results" / "83_revision_feedback.csv"
df.to_csv(csv_path, index=False)

print(csv_path)
````

````python
fig_path = repo / "figures" / "83_revision_feedback.png"

ax = df.plot.bar(
    x="revision_outcome",
    y="feedback_score",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Revision feedback")
ax.set_ylabel("Feedback score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````

## Result

Notebook 83 closes the feedback loop:

