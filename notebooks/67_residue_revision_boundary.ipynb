# 67 — Residue Revision Boundary

## Where is the boundary between useful revision and accumulated noise?

Notebook 61 examined accumulation.

Notebook 67 asks:

> When should accumulated residues be revised instead of preserved?

Outputs:

```text
results/67_residue_revision_boundary.csv
figures/67_residue_revision_boundary.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results", "figures", "reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    [1, 4.5, 0.5],
    [2, 4.0, 1.0],
    [3, 3.4, 1.8],
    [4, 2.8, 2.5],
    [5, 2.2, 3.1],
    [6, 1.7, 3.8],
]

df = pd.DataFrame(
    rows,
    columns=[
        "revision_cycle",
        "residue_value",
        "noise_pressure",
    ],
)

df["boundary_score"] = df["residue_value"] - df["noise_pressure"]
df["decision"] = df["boundary_score"].apply(
    lambda x: "preserve" if x > 0 else "revise"
)

df
````

````python
csv_path = repo / "results" / "67_residue_revision_boundary.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "67_residue_revision_boundary.png"

ax = df.plot(
    x="revision_cycle",
    y=["residue_value","noise_pressure","boundary_score"],
    figsize=(9,5),
)

ax.set_title("Residue revision boundary")
ax.set_ylabel("Score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````

## Result

Notebook 67 introduces a simple boundary:

```text
residue value
    -
noise pressure
```
