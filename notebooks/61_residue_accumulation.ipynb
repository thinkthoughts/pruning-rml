# 61 — Residue Accumulation

## When does accumulation become noise?

Notebook 59 examined stability.

Notebook 61 asks:

> At what point does accumulated structure stop helping and start increasing maintenance cost?

Outputs:

```text
results/61_residue_accumulation.csv
figures/61_residue_accumulation.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results", "figures", "reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    [1, 1.0, 0.1],
    [2, 2.0, 0.3],
    [3, 3.0, 0.7],
    [4, 3.5, 1.4],
    [5, 3.6, 2.2],
    [6, 3.4, 3.1],
]

df = pd.DataFrame(
    rows,
    columns=[
        "revision_cycle",
        "signal_strength",
        "maintenance_cost",
    ],
)

df["net_value"] = df["signal_strength"] - df["maintenance_cost"]
df
````

````python
csv_path = repo / "results" / "61_residue_accumulation.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "61_residue_accumulation.png"

ax = df.plot(
    x="revision_cycle",
    y=["signal_strength","maintenance_cost","net_value"],
    figsize=(9,5),
)

ax.set_title("Residue accumulation")
ax.set_ylabel("Score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
