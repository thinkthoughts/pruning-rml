# 103 — Real Pruning Experiments

## Moving beyond toy examples

This notebook identifies pruning-RML hypotheses that should be tested
against real benchmarks.

Outputs:

```text
results/103_real_pruning_experiments.csv
figures/103_real_pruning_experiments.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["residue preservation predicts accuracy", 5, "benchmark"],
    ["trigger score predicts pruning benefit", 5, "benchmark"],
    ["queue ordering improves efficiency", 4, "ablation"],
    ["policy selection reduces retraining cost", 4, "benchmark"],
    ["governance improves lifecycle outcomes", 2, "longitudinal"],
]

df = pd.DataFrame(
    rows,
    columns=["hypothesis","priority","experiment_type"]
)

df
````

````python
csv_path = repo / "results" / "103_real_pruning_experiments.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "103_real_pruning_experiments.png"

ax = df.plot.bar(
    x="hypothesis",
    y="priority",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Real pruning experiment priorities")
ax.set_ylabel("Priority")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
