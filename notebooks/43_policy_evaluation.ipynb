# 43 — Policy Evaluation

## Comparing pruning policies

Notebook 41 defined policies.

Notebook 43 asks:

> Which policy performs best under the same constraints?

Outputs:

```text
results/43_policy_evaluation.csv
figures/43_policy_evaluation.png
```

````python
from pathlib import Path
import os, sys, subprocess

REPO_URL = "https://github.com/thinkthoughts/pruning-rml.git"
REPO_NAME = "pruning-rml"

cwd = Path.cwd().resolve()

if cwd.name == REPO_NAME:
    repo = cwd
elif Path("/content/pruning-rml").exists():
    repo = Path("/content/pruning-rml")
elif (cwd / REPO_NAME).exists():
    repo = cwd / REPO_NAME
else:
    target = Path("/content") if Path("/content").exists() else cwd
    os.chdir(target)
    subprocess.run(["git","clone",REPO_URL], check=True)
    repo = target / REPO_NAME

os.chdir(repo)

src = repo / "src"
if str(src) not in sys.path:
    sys.path.insert(0, str(src))
````

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

def ensure_dirs(root, names=("results","figures","reports")):
    root = Path(root)
    out={}
    for n in names:
        p=root/n
        p.mkdir(parents=True, exist_ok=True)
        out[n]=p
    return out

outputs = ensure_dirs(repo)
results = outputs["results"]
figures = outputs["figures"]
````

````python
rows = [
    ["preserve_and_refine", 0.88, 0.80, 0.90],
    ["targeted_revision", 0.82, 0.85, 0.78],
    ["major_restructure", 0.70, 0.65, 0.72],
]

df = pd.DataFrame(
    rows,
    columns=[
        "policy",
        "capability_retention",
        "cost_efficiency",
        "adaptability",
    ],
)

df["overall_score"] = (
    df["capability_retention"]
    + df["cost_efficiency"]
    + df["adaptability"]
) / 3

df
````

````python
csv_path = results / "43_policy_evaluation.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = figures / "43_policy_evaluation.png"

ax = df.plot.bar(
    x="policy",
    y="overall_score",
    legend=False,
    figsize=(9,5),
)

ax.set_title("Policy evaluation")
ax.set_ylabel("Overall score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
