# 101 — Repo Retrospective

## What survives from the pruning-RML exploration?

This notebook reviews the ideas explored across the repo.

Outputs:

```text
results/101_repo_retrospective.csv
figures/101_repo_retrospective.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["residues", 5, "src module"],
    ["triggers", 5, "src module"],
    ["revision queue", 4, "src module"],
    ["policy evaluation", 4, "benchmark"],
    ["governance", 2, "documentation"],
    ["architecture", 3, "roadmap"],
]

df = pd.DataFrame(
    rows,
    columns=["idea","priority","promotion_target"]
)

df
````

````python
csv_path = repo / "results" / "101_repo_retrospective.csv"
df.to_csv(csv_path, index=False)
print(csv_path)
````

````python
fig_path = repo / "figures" / "101_repo_retrospective.png"

ax = df.plot.bar(
    x="idea",
    y="priority",
    legend=False,
    figsize=(9,5),
)

ax.set_title("Repo retrospective")
ax.set_ylabel("Priority")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````

## Result

The retrospective separates exploratory notebook ideas from candidates
for stable implementation.
