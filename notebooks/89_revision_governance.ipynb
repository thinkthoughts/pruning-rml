# 89 — Revision Governance

## Governing long-term revision decisions

Notebook 83 closed the feedback loop.

Notebook 89 asks:

> How should accumulated feedback influence long-term governance of revision, pruning, and reset decisions?

Outputs:

```text
results/89_revision_governance.csv
figures/89_revision_governance.png
```

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo = Path.cwd()

for candidate in [repo, Path("/content/pruning-rml")]:
    if candidate.exists():
        repo = candidate
        break

for name in ["results","figures","reports"]:
    (repo / name).mkdir(parents=True, exist_ok=True)
````

````python
rows = [
    ["local optimization", 0.90, 0.40, 0.30],
    ["periodic review", 0.85, 0.70, 0.60],
    ["policy oversight", 0.80, 0.85, 0.90],
    ["full lifecycle governance", 0.88, 0.95, 0.95],
]

df = pd.DataFrame(
    rows,
    columns=[
        "governance_model",
        "adaptability",
        "accountability",
        "long_term_stability",
    ],
)

df["governance_score"] = (
    df["adaptability"]
    + df["accountability"]
    + df["long_term_stability"]
) / 3

df
````

````python
csv_path = repo / "results" / "89_revision_governance.csv"
df.to_csv(csv_path, index=False)

print(csv_path)
````

````python
fig_path = repo / "figures" / "89_revision_governance.png"

ax = df.plot.bar(
    x="governance_model",
    y="governance_score",
    legend=False,
    figsize=(10,5),
)

ax.set_title("Revision governance")
ax.set_ylabel("Governance score")

plt.tight_layout()
plt.savefig(fig_path, dpi=160)
plt.show()

print(fig_path)
````
